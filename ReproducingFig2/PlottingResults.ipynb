```python
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns
import matplotlib as mpl
from matplotlib.gridspec import GridSpec
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['text.usetex'] = True
plt.rc('text.latex', preamble=r'\usepackage{amsmath,dsfont}')
```

```python
#loading data for active neurons
D=np.loadtxt('ActiveNodeDensity.txt')
# time
t=D[:,0]
# excitatory
NE=D[:,1]
# inhibitory
NI=D[:,2]


# loading data for active synapses
L=np.loadtxt('ActiveLinkDensity.txt')
# excitatory
LE=L[:,1]
# inhibitory
LI=L[:,2]
```

```python
# Plotting fraction of active excitatory and inhibitory neurons as a function of time
fig, ax=plt.subplots(figsize=(12,2), facecolor='white')
ax.set_facecolor('whitesmoke')
gs=GridSpec(8,15)

plt.plot(t,NE,'o--',color='C0', lw = 1.5, markersize=3.5 , label=r'$\rho_{\mathcal{E}}$')
plt.plot(t,NI,'^--',color='C3', lw = 1.5, markersize=3.5 , label=r'$\rho_{\mathcal{I}}$')
plt.xlabel('time',fontsize=14, rotation=0)
plt.ylabel('active density',fontsize=14, rotation=90)
plt.grid(color = 'white', linestyle = '-', linewidth = 2)
plt.legend()
```

```python
# Plotting fraction of active excitatory and inhibitory synapses as a function of time
fig, ax=plt.subplots(figsize=(12,2), facecolor='white')
ax.set_facecolor('whitesmoke')
gs=GridSpec(8,15)

plt.plot(t,LE,'s--',color='C2', lw = 1.5, markersize=3.5 , label=r'$\phi_{\mathds{E}}$')
plt.plot(t,LI,'*--',color='C4', lw = 1.5, markersize=3.5 , label=r'$\phi_{\mathds{I}}$')
plt.xlabel('time',fontsize=14, rotation=0)
plt.ylabel('active density',fontsize=14, rotation=90)
plt.grid(color = 'white', linestyle = '-', linewidth = 2)
plt.legend()
```

```python
# Plotting fraction of total active neurons and synapses as a function of time
fig, ax=plt.subplots(figsize=(12,2), facecolor='white')
ax.set_facecolor('whitesmoke')
gs=GridSpec(8,15)

Rho=NE+NI
plt.plot(t,Rho,'--',color='C9', lw = 1.5, markersize=3.5 , label=r'$\rho$')
plt.xlabel('time',fontsize=14, rotation=0)
plt.ylabel('active density',fontsize=14, rotation=90)
plt.grid(color = 'white', linestyle = ':', linewidth = 2)
plt.legend()

Phi=LE+LI
plt.plot(t,Phi,'-.',color='C7', lw = 1.5, markersize=3.5 , label=r'$\phi$')
plt.xlabel('time',fontsize=14, rotation=0)
plt.ylabel('active density',fontsize=14, rotation=90)
plt.grid(color = 'white', linestyle = '-', linewidth = 2)
plt.legend()
```

```python
# Plotting mean input current 
fig, ax=plt.subplots(figsize=(12,2), facecolor='white')
ax.set_facecolor('whitesmoke')
gs=GridSpec(8,15)

meanCur=LE-4*LI
plt.plot(t,meanCur,'v--',color='C5', lw = 1.5, markersize=3.5 , label=r'$\phi_{\mathds{E}}-4\phi_{\mathds{I}}$')
plt.xlabel('time',fontsize=14, rotation=0)
plt.ylabel('active density',fontsize=14, rotation=90)
plt.grid(color = 'white', linestyle = '-', linewidth = 2)
plt.legend()
```

```python
# Plotting normalized input current of a neuron chosen randomly 
C=np.loadtxt('NodeCurrent.txt')
# time
tt=C[:,0]
# current
nodeCur=C[:,1]/400  # the maximum input current of a neuron in the network = mean degree of Exc links = 400

fig, ax=plt.subplots(figsize=(12,2), facecolor='white')
ax.set_facecolor('whitesmoke')
gs=GridSpec(8,15)

plt.plot(tt,nodeCur, color='chocolate',  lw = 2, label=r'$\Delta_{in}^i$')
plt.xlabel('time',fontsize=14, rotation=0)
plt.ylabel('active density',fontsize=14, rotation=90)
plt.grid(color = 'white', linestyle = '-', linewidth = 2)
plt.legend()


# times in which the neuron fires
spikes=[]     
times=[]
for i in np.arange(len(nodeCur)):
    if (nodeCur[i]>=0.01):    # normalized threshold = 4/400
        times.append(i+1)
        spikes.append(nodeCur[i])
plt.plot(times,spikes, 'x', color='black',  markersize= 6)
plt.axhline(y = 0.008, color = 'C7', linestyle = ':')
```
